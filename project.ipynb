```python
from gurobipy import *
m = Model()

x1 = m.addVar(vtype=GRB.INTEGER)
x2 = m.addVar(vtype=GRB.INTEGER)
x3 = m.addVar(vtype=GRB.INTEGER)
x4 = m.addVar(vtype=GRB.INTEGER)
y1 = m.addVar(vtype=GRB.INTEGER)
y2 = m.addVar(vtype=GRB.INTEGER)
y3 = m.addVar(vtype=GRB.INTEGER)
y4 = m.addVar(vtype=GRB.INTEGER)
z1 = m.addVar(vtype=GRB.INTEGER)
z2 = m.addVar(vtype=GRB.INTEGER)
z3 = m.addVar(vtype=GRB.INTEGER)
z4= m.addVar(vtype=GRB.INTEGER)

PM = 4
ES = 1.25
ADS = 5
DSN = 3
DEV = 1
CW = 0.5
SN = 3

m.update

m.addConstr((1/4)*(x1+y1) <= PM)
m.addConstr((1/4)*(x2+y2) <= PM)
m.addConstr((1/4)*(x3+y3) <= PM)
m.addConstr((1/4)*(x4+y4) <= PM)


m.addConstr(1/6*(x1+y1) + 1/4*(x2+y2) + 1/8*(x3+y3) + 1/8*(x4+y4) <= PM)


m.addConstr((3/10)*(x1+y1) <= ES)
m.addConstr((3/10)*(x1+y1) <= ES)
m.addConstr((3/10)*(x2+y2) <= ES)
m.addConstr((3/10)*(x3+y3) <= ES)
m.addConstr((3/10)*(x4+y4) <= ES)


m.addConstr(1/16*(x2+y2) + 1/10*(x3+y3) + 1/10*(x4+y4) <= ES)


m.addConstr((1/4)*(x2+y2) <= SN)
m.addConstr((1/4)*(x3+y3) <= SN)
m.addConstr((1/4)*(x4+y4) <= SN)
m.addConstr((1/4)*(x1+y1) <= SN)

m.addConstr((1/4)*(x2+x3+x4+y2+y3+y4) <= SN)


m.addConstr((3/20)*(x1+y1) + (1/20)*z1 <= DSN)
m.addConstr((3/10)*(x1+y1) + (1/40)*z1 <= DSN)

m.addConstr((3/20)*(x2+y2) + (1/20)*z2 <= DSN)
m.addConstr((3/10)*(x2+y2) + (1/40)*z2 <= DSN)

m.addConstr((3/20)*(x3+y3) + (1/20)*z3 <= DSN)
m.addConstr((3/10)*(x3+y3) + (1/40)*z3 <= DSN)

m.addConstr((3/20)*(x4+y4) + (1/20)*z4 <= DSN)
m.addConstr((3/10)*(x4+y4) + (1/40)*z4 <= DSN)


m.addConstr(x1*(1/10) + x2*(1/5) + x3*(1/10) + x4*(1/5) 
            + y1*(1/10) + y2*(1/5) + y3*(1/10) + y4*(1/5)
            + z1*(1/16) + z2*(1/16) + z3*(1/16) + z4*(1/16) <= DSN)


m.addConstr((1/10)*(x1+y1) + (1/5)*z1 <= ADS)
m.addConstr((1/5)*(x1+y1) + (1/8)*z1 <= ADS)
m.addConstr((1/10)*(x2+y2) + (1/5)*z2 <= ADS)
m.addConstr((1/5)*(x2+y2) + (1/8)*z2 <= ADS)
m.addConstr((1/10)*(x3+y3) + (1/5)*z3 <= ADS)
m.addConstr((1/5)*(x3+y3) + (1/8)*z3 <= ADS)
m.addConstr((1/10)*(x4+y4) + (1/5)*z4 <= ADS)
m.addConstr((1/5)*(x4+y4) + (1/8)*z4 <= ADS)


m.addConstr(x1*(1/10) + x2*(1/5) + x3*(1/10) + x4*(1/5) 
            + y1*(1/10) + y2*(1/5) + y3*(1/10) + y4*(1/5) 
            + z1*(1/5) + z2*(1/8) + z4*(1/8) + z4*(1/8) <= ADS)


m.addConstr((3/40)*(x1+y1) + (1/10)*z1 <= DEV)
m.addConstr((1/40)*(x1+y1) <= DEV)
m.addConstr((3/40)*(x2+y2) + (1/10)*z2 <= DEV)
m.addConstr((1/40)*(x2+y2) <= DEV)
m.addConstr((3/40)*(x3+y3) + (1/10)*z3 <= DEV)
m.addConstr((1/40)*(x3+y3) <= DEV)
m.addConstr((3/40)*(x4+y4) + (1/10)*z4 <= DEV)
m.addConstr((1/40)*(x4+y4) <= DEV)


m.addConstr(x1*(3/40) + x2*(1/40) + x3*(1/40) 
            + y1*(3/40) + y2*(1/40) + y3*(1/40) 
            + z1*(1/10) <= DEV)


m.addConstr((3/40)*(x1+y1) <= CW)
m.addConstr((3/40)*(x2+y2) <= CW)
m.addConstr((3/40)*(x3+y3) <= CW)
m.addConstr((3/40)*(x4+y4) <= CW)

m.addConstr(x1*(3/40) + x2*(1/20) 
            + y1*(3/40) + y2*(1/20) <= CW)

m.addConstr(y1 >= 2)
m.addConstr(y2 >= 2)
m.addConstr(y3 >= 2)
m.addConstr(y4 >= 2)

m.addConstr(z1 <= 3)
m.addConstr(z2 <= 3)
m.addConstr(z3 <= 3)
m.addConstr(z4 <= 3)



m.setObjective(18243.5*(x1+x2+x3+x4)+17737*(y1+y2+y3+y4)+5000*(z1+z2+z3+z4), GRB.MAXIMIZE)
m.optimize()
print(x1.x, y1.x, z1.x, x2.x, y2.x, z2.x, x3.x, y3.x, z3.x, x4.x, y4.x,z4.x)
```
